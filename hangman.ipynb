```python
#第十章 Hangman猜谜游戏

def hangman(zongli):
    wrong=0 #从1开始数
    stages=[
      '       ',
      '|      ',
      '|   |   ',
      '|   0   ',
      '|  /|\  ',
      '|  / \  ',
      '|       ',
       ]
    rletters=list(zongli)
    board=['__'] * len(zongli)
    win=False
    print('Welcom to Hangman')
    
    while wrong < len(stages)-1:
        print('\n')    # '\n'是空白行
        msg='Gess a letter:'  #msg:提示
        char=input(msg)
        if char in rletters:
            cind=rletters.index(char)  #这三行很重要!cind是个索引数字
            board[cind]=char   # 更新board
            rletters[cind]='$' #更新rletters, 避免系统出错用$
        else:
            wrong += 1
        
        print(''.join(board)) #用''焊接
        e=wrong+1
        print('\n'.join(stages[0:e]))
        if '__'not in board:
            print('you win!')
            print(''.join(board))
            win=True
            break
    if not win:
        print('\n'.join(stages[0:wrong]))
        print('you lose!,it was {}.'.format(zongli))

hangman('zongli')          
            
```

```python

#定义吊死你函数 优化版

def 吊死你(nicolas):
    错误 = 0
    死亡进程 = ['',          
    '         ',
    '|        ',          
    '|    |   ',
    '|    0  ',          
    '|   /|\ ',
    '|   / \ ',          
    '|        ']
    
    名字 = list('nicolas') 
    成功进程 = ['__'] * len(nicolas)
    成功 = False
    print('欢迎来到死亡游戏吊死你:')

# 循环   
    while 错误 < len(死亡进程):  #可-1      
        print('\n')
        输入=input('请猜一个字母?:')
        if 输入 in 名字:
            索引数=名字.index(输入)
            成功进程[索引数]=输入
            名字[索引数]='$'
        else:
            错误 += 1
        print(''.join(成功进程))
        e=错误+1
        print('\n'.join(死亡进程[0:e]))

#   结局时刻     
        
        if '__' not in 成功进程:
            print('你成功了!')
            print(''.join(成功进程))
            成功=True
            break
    if not 成功:
        print('\n'.join(死亡进程[0:错误]))
        print('你输了笨笨!名字叫{}.'.format(nicolas))
        
吊死你('nicolas')
```
